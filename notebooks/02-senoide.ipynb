$x[n] = x(n T_s)$

$x(t) = \cos(2 \pi f_0 t)$

$x[n] = \cos(2 \pi f_0 n T_s)$

```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Audio
```

```python
fs = 44100
Ts = 1/fs
f0 = 500
n = np.arange(44100*5)
#n = np.arange(44100)
#xn = np.cos(2 * np.pi * f0 * n * Ts)
#xn = np.cos(2 * np.pi * f0 * n * Ts) * 0.1
#xn = np.random.rand(len(n)) * 0.1
xn = np.cos(2 * np.pi * 2 * f0 * n * Ts) * 0.1
```

```python
plt.figure()
plt.plot(n[0:200]*Ts, xn[0:200], label='Sinal amostrado')
plt.ylabel('Magnitude')
plt.xlabel('Tempo (s)')
#plt.legend()
plt.show()
```

```python
Audio(data=xn, rate=fs)
```
