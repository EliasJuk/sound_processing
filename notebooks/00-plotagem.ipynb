```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot([0, 1, 2, 3, 4, 5, 6, 7, 8], [0, 1, -1, 1.5, -1, 2, -1.5, 1.5, 0])
plt.show()
```

```python
plt.figure()
plt.plot([0, 1, 2, 3, 4, 5, 6, 7, 8], [0, 1, -1, 1.5, -1, 2, -1.5, 1.5, 0], label="Wave-1")
plt.plot([0, 1, 2, 3, 4, 5, 6, 7, 8], [1, -1.5, 2, -2, 1, -1, 1.5, -1, 2], label="Wave-2")
plt.ylabel('Magnitude')
plt.xlabel('Mediana')
plt.legend()
plt.show()
```
